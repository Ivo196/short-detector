### Data Acquisition

The yfinance library is used to download interval data and save it in raw_data for the next step, data_indicators, where various indicators are calculated.

```python
import yfinance as yf
import pandas as pd

# 1. Configuración
TICKER = "ETH-USD"
INTERVALO = "1d" 

print(f"Descargando histórico diario para {TICKER}...")

# 2. Descarga (period="max" trae todo el historial disponible)
df_raw = yf.download(TICKER, period="max", interval=INTERVALO, multi_level_index=False)

# 3. Guardar CSV Crudo
nombre_archivo_raw = f"ethereum_{INTERVALO}_raw.csv"

df_raw.to_csv(f"raw_data/{nombre_archivo_raw}")

print(f"✅ Datos guardados exitosamente en: {nombre_archivo_raw}")
print(f"Total de registros: {len(df_raw)}")
print(df_raw.tail(3))
```
